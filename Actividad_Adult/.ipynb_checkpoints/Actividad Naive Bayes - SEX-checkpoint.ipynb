# Naive Bayes 

```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import OneHotEncoder

#Se cargan y observan los datos
data_frame = pd.read_csv("3_2_3_adult.csv")
data_frame
```

```python
#Se observan los tipos de datos existentes
#Se deciden las principales variables a utilizar, como lo son edad,sexo, ingreso, numero de educación,etc.
data_frame.dtypes
```

```python
data_frame.describe()
```

Luego de analizar las medidas que se le realizaron a la muestra se puede concluir que hay un gran desbalance en la distribución de los datos, se puede ver claramente en la diferencia existente entre el segundo y tercer cuartil en la columa de capital ganado, en la que la que mas del 75% de los registros se encuentra bajo la media.

```python
#Se agrupan los datos con respecto a la edad con el fin de posteriormente crear intervalos para dicha columna
print(data_frame.groupby('relationship').size())
```

```python
#Se crea un intervalo con los digitos utilizados para la edad

bins = [0, 30, 50, 70, 90,150]
#Se le asigna un nombre a cada intervalo
groups = ['-30','31-50','51-70','71-90','+90']
#Se cargan los datos en una nueva tabla llamada 'edad'
data_frame['edad'] = pd.cut(data_frame['age'],bins , labels = groups)

#Se reemplazan los valores masculinos y femeninos por un 0 y 1 respectivamente, para posteriormente
#utilizarlos como un eje principal, actualmente quedan definidos como un boolean, es decir un si y no
data_frame["sex"] = data_frame["sex"].replace("Female",0)
data_frame["sex"] = data_frame["sex"].replace("Male",1)

#Se hace lo mismo que se hizo con sexo pero esta vez con la variable de ingresos, dividiendolos en las
#personas que cuenten con un ingreso mayor o menor a 50K
data_frame["income"] = data_frame["income"].replace("<=50K",0)
data_frame["income"] = data_frame["income"].replace(">50K",1)

#Se realiza el onehotencoding en la tabla relationship para posteriormente actualizar el dataframe con
#las columnas nuevas


#Se crea una condicion en la cual se clasificarán por la nacionalidad del registrado, inicialmente
#la distribución de las nacionalidades era muy dispareja, por lo cual se redujo a 2 posibilidades, la
#que corresponde a los estadounidenses y a los que no lo son. Equiparando lo mas posible los registros 
data_frame['native.country'] = data_frame['native.country'].replace("United-States","American")
data_frame['native.country'] = data_frame['native.country'].mask(data_frame['native.country']!='American','Non-American') 

data_frame['relationship'] = data_frame['relationship'].replace("Not-in-family","Not-in-family")
data_frame['relationship'] = data_frame['relationship'].replace("Unmarried","Not-in-family")
data_frame['relationship'] = data_frame['relationship'].mask(data_frame['relationship']!="Not-in-family",'In-family') 


#Se dividen los tipos de trabajo en privado, y no privado. La razon de esto es parecida a la mencionada                                                      ) 
#en el punto anterior, basicamente se hizo con el fin de equiparar las categorias y no dejarlas desnalanceadas
data_frame['workclass'] = data_frame['workclass'].mask(data_frame['workclass']!='Private','Non-private')

```

```python
#Se realiza el onehotencoder de forma manual para explicar mejor su funcionamiento
ohe = OneHotEncoder()
print(ohe)
```

```python
#Primero tomamos la categoria a usar, en este caso corresponde a la clase de trabajo que anteriormente
#editamos en privadas y no privadas, acto seguido transformamos ambas variables en un array, para
#el registro de los datos
ohe.fit_transform(data_frame[["workclass"]]).toarray()
arry = ohe.fit_transform(data_frame[["workclass"]]).toarray()
```

```python
#Se especifica que el valor de los labels a traves de las categorias, en este caso es privado y no privado
ohe.categories_
labels = ohe.categories_
```

```python
#Se crea la matriz, actualizando el valor de labels, que anteriormente solo era las categorias, ahora 
#se vuelve una matriz
np.array(labels).ravel()
labels = np.array(labels).ravel()
print(labels)
```

```python
#se crea la matriz final, dejando como resultado los registros distribuidos dependiendo de la variable
#a la que correspondan, privadas o publicas, dejando un valor de 1 en los que sea true
pd.DataFrame(arry, columns = labels)
matriz = pd.DataFrame(arry, columns = labels)
```

```python
#Se cargan los datos previamente expuestos en el dataframe para que se puedan ver los registros en la tala
pd.concat([data_frame, matriz], axis = 1)
data_frame = pd.concat([data_frame, matriz], axis = 1)
```

Se realiza un grafico de torta con el fin de saber la distribución de los datos en el dataframe. Todo se hace en base al sexo, pudiendo identificar una superioridad en la cantidad de registros masculinos
con un %

```python
import matplotlib.pyplot as plt

clases = np.array([data_frame[data_frame.sex == 0].shape[0], data_frame[data_frame.sex == 1].shape[0]])
labels = [str(round(x * 1.0 / clases.sum() * 100.0, 2)) + '%' for x in clases]

labels[0] = 'Mujeres\n ' + labels[0]
labels[1] = 'Hombres\n ' + labels[1]

plt.pie(clases, labels=labels)
plt.title('Porcentaje de mujeres y hombres')
plt.show()
```

```python
#Se realiza otro dtypes para verificar el tipo de datos a los que corresponden las columnas agregadas
#a traves del onehotencoding, mostrando todos el mismo tipo de dato, que corresponde a un float64
data_frame.dtypes
```

Se crea un segundo dataframe el cual eliminará las columnas object, y las que se consideren innecesarias en el futuro, como lo puede ser el capital perdido o el ganado, puesto que por la distribución irregular que tiene, de momento se dejará excluida para analizarlos posteriormente

```python
dropped_dframe = data_frame.drop(['workclass','age','edad','income','workclass','education','education.num','marital.status', 'occupation','relationship','race','native.country','fnlwgt'], axis=1)
dropped_dframe.dtypes
```

```python
#Se muestra una variedad de graficos que muestra el comportamiento de todas las variables x con respecto
#a nuestra variable y(sexo)
import seaborn as sns

columnas = np.array(dropped_dframe.columns)

for col in columnas[:-1]:
    fig, ax =plt.subplots(1, 2, figsize=(20, 6))
    fig.suptitle(col, fontsize=18)
    sns.distplot(data_frame[col], ax=ax[0], kde=False)
    data_frame[[col]+['sex']].plot.scatter(x=col, y='sex', ax=ax[1])
    plt.show()

```

```python
#Se determinan los valores de x e y para su graficación mas adelante
X = dropped_dframe.drop(["sex"], axis=1)
y = dropped_dframe.sex
```

```python
#Se prepara el modelo de Naive Bayes para el testeo de los datos
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=25)

modelo = GaussianNB()

modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)

```

```python
#Se evalua el proyecto
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)

# Nos muestra la precision del modelo
from sklearn.metrics import precision_score
precision = precision_score(y_test, y_pred)
print('Precisión del modelo:', precision)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
print('Accurancy del modelo:',accuracy)
print('F1 score del modelo:',f1_score(y_test, y_pred))
print('Recall del modelo:',recall_score(y_test, y_pred))
```

```python
from sklearn.feature_selection import SelectKBest

X=dropped_dframe.drop(['sex'], axis=1)
y=dropped_dframe['sex']
best=SelectKBest(k='all')
X_new = best.fit_transform(X, y)
X_new.shape
selected = best.get_support(indices=True)
print(X.columns[selected])
```
